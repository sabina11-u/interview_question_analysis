# Software Engineering Interview Questions Analysis
### A data analysis project using Python and Pandas
**Dataset:** 9,000 software engineering interview questions
**Tools:** Python, Pandas, Matplotlib, Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv('new_interview_questions.csv')
print("Dataset loaded!")
print("Shape:", df.shape)
```

## Step 1: Exploring the Dataset
Let's take a first look at the data.

```python
# Show first 5 rows
df.head()
```

```python
# Check columns, data types and missing values
print("Columns:", df.columns.tolist())
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Summary statistics
print("Total Questions:", len(df))
print("\nCategories:")
print(df['Category'].value_counts())
print("\nDifficulty Levels:")
print(df['Difficulty'].value_counts())
```

## Step 2: Data Visualization
Now let's create charts to understand the data better.

```python
plt.figure(figsize=(12, 6))
df['Category'].value_counts().plot(kind='barh', color='steelblue')
plt.title('Number of Interview Questions by Category', fontsize=16)
plt.xlabel('Number of Questions')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
cross_tab = pd.crosstab(df['Category'], df['Difficulty'])
cross_tab.plot(
    kind='bar',
    stacked=True,
    color=['#e74c3c', '#2ecc71', '#f39c12'],
    figsize=(14, 7)
)
plt.title('Difficulty Breakdown by Category', fontsize=16)
plt.xlabel('Category')
plt.ylabel('Number of Questions')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Difficulty')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 7))
df['Difficulty'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['#2ecc71', '#f39c12', '#e74c3c'],
    startangle=90
)
plt.title('Question Difficulty Distribution', fontsize=16)
plt.ylabel('')
plt.show()
```

## Step 3: Key Insights
Here are the main findings from this analysis.

```python
total = len(df)
top_category = df['Category'].value_counts().index[0]
top_count = df['Category'].value_counts().values[0]
top_difficulty = df['Difficulty'].value_counts().index[0]
hard_category = df[df['Difficulty']=='Hard']['Category'].value_counts().index[0]

print("=" * 50)
print("KEY INSIGHTS")
print("=" * 50)
print(f"1. Total questions in dataset: {total}")
print(f"2. Most common category: {top_category} ({top_count} questions)")
print(f"3. Most common difficulty: {top_difficulty}")
print(f"4. Category with most Hard questions: {hard_category}")
print("5. Medium difficulty dominates, showing companies")
print("   prefer moderately challenging interview questions.")
print("=" * 50)
```
